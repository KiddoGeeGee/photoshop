```python
import cv2
import numpy as np
from PIL import Image
import sys

sys.path.append("../")
import utils.roop as roop
from utils.utils import convert_to_b64_list

sys.path.append("../../../git_data/CodeFormer/")
from app import inference_app


source_img = convert_to_b64_list([Image.open("jjl.jpg")])[0]
target_img = convert_to_b64_list([Image.open("zly.jpg")])[0]
roop_res = roop.main(source_img,target_img, roop.face_swapper)
scale = min(roop_res.size)//512
roop_res = roop_res.resize((roop_res.size[0]//scale, roop_res.size[1]//scale))
roop_res = cv2.cvtColor(np.array(roop_res),cv2.COLOR_RGB2BGR)
res = inference_app(roop_res,
                    background_enhance=False,
                    face_upsample=False,
                    upscale=1,
                    codeformer_fidelity=1,
                    )
res
```
